```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
df.info()
```

```python
# Change diagnosis column to 1 if M and 0 if B
df = df.replace('M', 1)
df = df.replace('B', 0)
```

```python
df.describe()
```

```python
X, y = df.drop(columns=['diagnosis', 'id']), df['diagnosis']
```

```python
# Set up the subplot grid (10 rows, 3 columns)
fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(16, 32))
sns.set_theme(style='darkgrid')

# Flatten axes to make indexing easier
axes = axes.flatten()

# Plot a boxplot for each feature in the dataset (excluding id's, and diagnosis)
for i, column in enumerate(X.columns):
    sns.boxplot(x=X[column], ax=axes[i], color='deepskyblue')
    axes[i].set_title(f'Boxplot for {column}')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculate Q1 and Q2 for each column1
Q1s = df.quantile(0.25)
Q3s = df.quantile(0.75)

# Calculate the lower and upper bounds for each column
IQRs = Q3s - Q1s
lower_bound = Q1s - 1.5 * IQRs
upper_bound = Q3s + 1.5 * IQRs

# Keep only rows where all columns are within bounds
within_bounds = (df >= lower_bound) & (df <= upper_bound)
df = df[within_bounds.all(axis=1)]
```

```python
X, y = df.drop(columns=['id', 'diagnosis']), df['diagnosis']
```

```python
# Standardize feature values:
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# Create and train svm model
svm_linear = svm.SVC(kernel = 'linear', random_state=42)
svm_linear.fit(X_train, y_train)

# Evaluate svm model on testing and training data
y_pred_train = svm_linear.predict(X_train)
y_pred_test = svm_linear.predict(X_test)
accuracy_train = accuracy_score(y_pred_train, y_train)
accuracy_test = accuracy_score(y_pred_test, y_test)
```

```python
print(f'Train accuracy: {accuracy_train}')
print(f'Test accuracy: {accuracy_test}')
```
